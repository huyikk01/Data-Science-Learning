* **Box Plot**: Phương pháp để mô tả bằng đồ thị các nhóm dữ liệu số thông qua các phần tử của chúng
* **Join Plot**

### Distribution Plot
* **Univariate distribution of observations**

```python
tips_df=sns.load_dataset('tips')
tips_df.head()
```

```python
## Styling
sns.set_theme() #Apply default
```

```python
sns.histplot(tips_df["total_bill"]);
```

```python
tips_df["total_bill"].value_counts().sort_values(ascending=False)
```

### KDE Plot

```python

sns.kdeplot(data=tips_df['total_bill']);
```

### Displot


```python
sns.displot(data=tips_df,x="total_bill",col="time",kde=True);
```

## Bar Plots

```python
tips_df.head()
```

```python
sns.barplot(data=tips_df,x="sex",y="tip",estimator=np.mean);
```

## Count Plot

```python
tips_df["sex"].value_counts()
```

```python
sns.countplot(data=tips_df, x="sex");
```

## Box Plot

```python
sns.boxplot(data=tips_df,x="day",y="total_bill",hue="sex");
plt.legend(loc=0);
```

### Facet Grid

```python
tips_df.head()
```

```python
tips_fg=sns.FacetGrid(data=tips_df,row="smoker",col="time");
tips_fg.map(sns.scatterplot,'total_bill','tip');
```

```python
new_fg=sns.FacetGrid(data=tips_df,col="sex",col_order=["Female","Male"],hue="smoker",height=4,aspect=1.4)
new_fg.map(sns.scatterplot,'total_bill','tip',s=100);
new_fg.add_legend();
```

### Join Plot

```python
penguin_df=sns.load_dataset('penguins')
penguin_df.head()
```

```python
sns.jointplot(data=penguin_df,x="flipper_length_mm",y="bill_length_mm");
```

### HeatMap

```python
flights_df=sns.load_dataset("flights")
```

```python
flights_df.head()
```

```python
flights=pd.pivot_table(flights_df,index='month',columns='year',values='passengers',observed=False);
```

```python
flights
```

```python
sns.heatmap(data=flights,cmap='Blues')
```
