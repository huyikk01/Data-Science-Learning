### Pyplot API vs OO API

```python
x=np.linspace(0,10,1000)
plt.plot(x,np.sin(x),linestyle='dashed',label="sin(x)");
plt.plot(x,np.cos(x),color='red',label="cos(x)");
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Sin(x) and Cos(x)")
plt.xlim([0,8])
plt.legend();
```

### OO API

```python
x=[1,2,3,4]
y=[11,22,33,44]
fig,ax=plt.subplots(figsize=(5,5)) #Figure size = Width& Height 
#Plot the data
ax.plot(x,y); 
ax.set(title="A simple Plot",
xlabel="x-axis",
ylabel="y-axis");
```

## Scatter

```python
plt.scatter(x,np.exp(x)); ##Pyplot
```

```python
##OO API
rng=np.random.RandomState(0)
x=rng.randn(100)
y=rng.randn(100)

colors=rng.rand(100)
sizes=1000*rng.rand(100)

fig, ax=plt.subplots()
ax.scatter(x,y,s=sizes,c=colors,cmap='viridis',alpha=0.3);
```

### Bar

```python
soft_drink_prices={"Coke":10,
                    "Pepsi":12,
                    "Sprite":8}
fig,ax = plt.subplots()
ax.bar(soft_drink_prices.keys(), soft_drink_prices.values());
ax.set(title="Shopping",ylabel="Price($)",xlabel="Drink")
```

### Histogram

```python
student_height=np.random.normal(170,10,250)
```

```python
plt.hist(student_height);
```

### Subplots

```python
#Option 1:
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
#Plot data
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(soft_drink_prices.keys(), soft_drink_prices.values());
ax4.hist(student_height,bins=20);
```

```python
## Option 2
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(soft_drink_prices.keys(), soft_drink_prices.values());
ax[1,1].hist(student_height,bins=20);
```
