## Project Data Visualization - California Cities

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cities=pd.read_csv('./california_cities.csv')
cities.head()
```

```python
#Extract Latd and Longd
lat,lon= cities['latd'],cities['longd']
population,area= cities['population_total'],cities['area_total_km2']
plt.style.use('seaborn-v0_8')
plt.figure(figsize=(8,6));
#Plot
plt.scatter(lon, lat,
            c=np.log10(population),cmap='viridis',
           s=area,linewidths=0,alpha=0.8);
plt.axis('equal');
plt.xlabel('longtitude');
plt.ylabel('latitude');
plt.colorbar(label='log$_{10}$(population)');
#Create a legend for cities' size
area_range=[50,100,300,500]
for area in area_range:
    plt.scatter([],[],s=area,c='k',alpha=0.6,
                label=str(area)+'km$^2$');
    plt.legend(labelspacing=1,title='City Area')
plt.title('California Cities: Population and Area Distribution');
```
