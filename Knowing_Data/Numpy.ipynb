### Array Concatenation and Splitting

```python
x=np.array([1,2,3])
y=np.array([2,3,4])
```

```python
np.concatenate((x,y))
```

```python
grid=np.array([[1,2,3],[4,5,6]])
grid
```

```python
np.concatenate((grid,grid),axis=1)
```

```python
#vstack
x=np.array([1,2,3])
grid=np.array([[2,3,4],[4,4,4]])
```

```python
np.vstack((x,grid))
```

```python
# hstack
y=np.array([[99],
            [99]])
np.hstack((y,grid))
```
